# 2) Найдите топ 100 самых лучших и самых худших фильмов (согласно среднему рейтингу, который прийдется посчитать)
 

```python
ratings=pd.read_csv('../../../lessons/24-11-2021/homework/data/ratings.csv')
ratings.nunique()
```

```python
ratings=ratings.rename(columns={'movieId':'id'})
ratings.dtypes
```

```python
movies_metadata=movies_metadata[movies_metadata['id'].apply(lambda x: x.isdigit())]
movies_metadata['id']=movies_metadata['id'].apply(int)
movies_metadata.columns
```

```python
all_merged=ratings.merge(movies_metadata,on=["id"])
all_merged
```

```python
all_merged.nunique()
```

```python
all_merged['rating'].shape, all_merged['rating'].dropna().shape
```

```python
ratings_group=all_merged.groupby('id').agg({'rating':['mean', 'count']})

ratings_group
```

```python
ratings_group.columns = ['mean_ratings', 'count_reviews']
```

```python
ratings_group
```

```python
ratings_group = ratings_group[ratings_group['count_reviews'] > 99]
```

```python
ratings_bad100=ratings_group.sort_values(by='mean_ratings',ascending=True)
ratings_bad100=ratings_bad100.head(100)

ratings_bad100 = ratings_bad100[0:].reset_index()
ratings_bad100
```
